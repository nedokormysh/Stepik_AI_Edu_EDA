# 1_3_10

Числа Фибоначчи

Напишите программу, которая выводит n-ое число ряда Фибоначчи.

Ввод
На вход подается строка, состоящая из одного неотрицательного целого числа (
$\overline{n}$)

Вывод
В единственной строке выводится значение n-ого числа Фибоначчи ($F_n$)

Замечание

Числа Фибоначчи — элементы числовой последовательности в которой первые два числа равны 0 и 1, а каждое последующее число равно сумме двух предыдущих чисел:
0, 1, 1, 2, 3, 5, 8, 13 ...

Будем считать, что $F_0 = 0$, $F_1 = 1$, $F_2 = 1$, ..., $F_n = F_{n-1} + F_{n-2}$

```python
def solution(n: int) -> int:
    # put your python code here

    fib1 = fib2 = 1

    if n == 0:
        return 0
    else:
      for i in range(2, n):
        #  print('Hello2')
         fib1, fib2 = fib2, fib1 + fib2
      return fib2

print(solution(int(input())))
```

Для начала давайте напомним как работают права доступа пользователей. Есть три вида разрешений. Они определяют права пользователя на три действия: чтение, запись и выполнение. В Linux эти действия обозначаются вот так:

R — read (чтение): право просматривать содержимое файла
W — write (запись): право изменять содержимое файла
X — execute (выполнение): право запускать файл, если это программа или скрипт

**Вывод**
Программа для каждого запроса должна будет возвращать OK если над файлом выполняется допустимая операция, или же Access denied, если операция недопустима

**Sample Output:**
